```python
df = pd.read_csv('Data/business_restaurant.csv')
df.head()
```

```python
df['categories'] = df['categories'].apply(lambda x: x.split(', '))
df.head()
```

```python
df2= pd.read_csv("Data/restaurant_photos.csv",lineterminator='\n')
```

```python
df2.head()
```

```python
df2['categories'] = ""
df2.head()
```

```python
# Import pandas and tqdm (if you're using tqdm for progress tracking in other parts of your code)
import pandas as pd
from tqdm.auto import tqdm

# Assuming 'df' and 'df2' are your pandas DataFrames and both have a 'business_id' column
# First, ensure that 'business_id' is the index for both DataFrames to speed up lookup
df.set_index('business_id', inplace=True)
df2.set_index('business_id', inplace=True)

# Use .loc to update 'categories' in df2 based on the index match
# tqdm can be used to show progress on operations that can be monitored, but in this case, the merge operation is atomic and efficient, so tqdm might not be necessary
df2['categories'] = df2.index.map(df['categories'])

# If you need to reset the index after the operation
df.reset_index(inplace=True)
df2.reset_index(inplace=True)
```
